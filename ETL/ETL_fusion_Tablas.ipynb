```python
# Cambiar el nombre de la columna 'shop_id' a 'id_shop' en df_tabla_dia_horarios_GOOGLE
df_tabla_dia_horarios_GOOGLE.rename(columns={'shop_id': 'id_shop'}, inplace=True)
df_tabla_dia_horarios_YELP.drop(columns=['id_yelp'], inplace=True)

# Agregar una columna 'id_time' incremental al DataFrame df_tabla_dia_horarios_YELP
df_tabla_dia_horarios_YELP.insert(0, 'id_time', range(1, len(df_tabla_dia_horarios_YELP) + 1))

# Agregar una columna 'id_time' incremental al DataFrame df_tabla_dia_horarios_GOOGLE
df_tabla_dia_horarios_GOOGLE.insert(0, 'id_time', range(len(df_tabla_dia_horarios_YELP) + 1, len(df_tabla_dia_horarios_YELP) + len(df_tabla_dia_horarios_GOOGLE) + 1))

# Unir los dos DataFrames en uno solo usando el método concat
frames = [df_tabla_dia_horarios_YELP, df_tabla_dia_horarios_GOOGLE]
day_time_Table = pd.concat(frames, ignore_index=True)

# day_time_Table ahora contiene la unión de los dos DataFrames con la columna 'id_time' como valor único incremental y el nombre de columna modificado
```

```python
df_tabla_Ubicacion_YELP = pd.read_csv('C:/Users/PC/Dropbox/HENRY/DATA-FT-12/Lab 03 - PF/DiagramaER/Ubicacion_Yelp_shops.csv')
df_tabla_Ubicacion_GOOGLE = pd.read_csv('C:/Users/PC/Dropbox/HENRY/DATA-FT-12/Lab 03 - PF/DiagramaER/location_g.csv', sep=';')
```

```python
df_tabla_Ubicacion_YELP.info()
```

```python
df_tabla_Ubicacion_GOOGLE.info()
```

```python
# Identificar los valores que generan el error
problematic_values = df_tabla_Ubicacion_GOOGLE.loc[~df_tabla_Ubicacion_GOOGLE['postal_code'].astype(str).str.isnumeric(), 'postal_code']

problematic_values
```

```python
# Identificar los valores no numéricos en la columna 'postal_code'
non_numeric_values = df_tabla_Ubicacion_GOOGLE[~df_tabla_Ubicacion_GOOGLE['postal_code'].str.isnumeric()]

non_numeric_values
```

```python
# Convertir el Dtype de la columna 'postal_code' a int64
df_tabla_Ubicacion_GOOGLE['postal_code'] = df_tabla_Ubicacion_GOOGLE['postal_code'].astype('Int64')
df_tabla_Ubicacion_GOOGLE.info()
```

```python
df_tabla_Ubicacion_YELP.sort_values(by='postal_code', inplace=True)
df_tabla_Ubicacion_YELP
```

```python
df_tabla_Ubicacion_GOOGLE.sort_values(by='postal_code', inplace=True)
df_tabla_Ubicacion_GOOGLE
```

```python
# Cambiar el nombre del campo 'adress' a 'address' en df_tabla_Ubicacion_GOOGLE
df_tabla_Ubicacion_GOOGLE.rename(columns={'adress': 'address'}, inplace=True)

# Unir los dos DataFrames en uno solo llamado df_location_Table
df_location_Table = pd.concat([df_tabla_Ubicacion_YELP, df_tabla_Ubicacion_GOOGLE], ignore_index=True)

# Ordenar el DataFrame por el campo postal_code
df_location_Table.sort_values(by='postal_code', inplace=True)

# df_location_Table ahora contiene la unión de los dos DataFrames con los campos renombrados y está ordenado por postal_code
```
